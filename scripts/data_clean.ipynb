```python
import pandas as pd
```

```python
df = pd.read_csv("03_Library Systembook.csv")
```

```python
df = df.dropna(how ='all')
```

```python
df.rename(columns={'Id': 'id', 'Books': 'book_title', 'Book checkout': 'checkout_date', 'Book Returned': 'return_date',
                   'Days allowed to borrow': 'time_allowed_to_barrow', 'Customer ID': 'customer_id' }, inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df['id'] = df['id'].astype(str)
df["book_title"] = df["book_title"].astype(str)
df['customer_id'] = df['customer_id'].astype(str)
```

```python
#standardise book title
def stadardise_book_title(title):
    if pd.isna(title):
        return title
    
    small_words = {"and", "the", "or", "an", "a"}
    words = title.lower().split()
    cleaned_words = []

    for i, word in enumerate(words):
        if i ==0 or word not in small_words:
            cleaned_words.append(word.capitalize())
        else:
            cleaned_words.append(word)
    
    return " ". join(cleaned_words)

df["book_title"] = df["book_title"].apply(stadardise_book_title)
```

```python
df['checkout_date'] = pd.to_datetime(df['checkout_date'], errors='coerce', dayfirst=True)
df['return_date'] = pd.to_datetime(df['return_date'], errors='coerce', dayfirst=True)
checkout = df["checkout_date"]
returned = df["return_date"]

borrow_time = (returned - checkout)
print(borrow_time)
```
